```python
import numpy as np
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
columns = ['Age', 'Salary', 'Profession']
data = [
    [23, 30000, 'Software Programmer'],
    [25, 35000, 'Software Programmer'],
    [30, 40000, 'System Analyst'],
    [35, 25000, 'Software Developer'],
    [40, 20000, 'HR Admin'],
    [30, 70000, 'Manager'],
    [35, 50000, 'Software Programmer'],
    [20, 30000, 'Software Programmer'],
]

df = pd.DataFrame(data = data, columns = columns)
df.head()
```

# Custom MinMax Scaler

```python
class myScaler(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self, X = None):
        return self
    
    def transform(self, X, y = None):
        minim = np.min(X)
        maxm = np.max(X)
        Z = (X-minim)/(maxm-minim)
        return Z
    
```

```python
mms = myScaler()
mms.fit_transform(df[['Age', 'Salary']])
```

# Custom Standard Scaler

```python
class myStandardScaler(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self, X = None):
        return self
    
    def transform(self, X, y = None):
        mean = np.mean(X)
        std = np.std(X)
        Z = (X-mean)/std
        return Z
```

```python
mms = myStandardScaler()
mms.fit_transform(df[['Age', 'Salary']])
```

```python
from sklearn.preprocessing import StandardScaler
df2 = df.copy()
ss = StandardScaler()
df2[['Age', 'Salary']] = ss.fit_transform(df2[['Age', 'Salary']])
```
